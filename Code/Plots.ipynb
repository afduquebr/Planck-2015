```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Frequencies of Planck
f = np. array([30, 44, 70, 100, 143, 217, 353, 545, 857]) # [GHz]
```

```python
def Synchrotron(f):
    return 2747547.1 * pow(f, -3.11)
```

```python
figure, axis = plt.subplots()
axis.loglog(f, Synchrotron(f), color='g', label='A')
axis.set_title("Synchrotron")
axis.set_xlabel("Frequency (GHz)")
axis.set_ylabel(r"Brightness temperature ($\mu$K)")
axis.legend()
axis.plot()
```

```python
def Dust(f, Bd, Td):
    Ad = 100
    f0 = 545
    kB = 1.380649 * pow(10, -23)
    h = 6.62607015 * pow(10, -34)
    gamma = h / (kB * Td) * pow(10, 9)
    return Ad * pow(f/f0,Bd+1) * ((math.exp(gamma*f0)-1)/(np.exp(gamma*f)-1))
```

```python
figure, axis = plt.subplots()
axis.loglog(f, Dust(f, 1.0, 21), color='g', label='A')
axis.loglog(f, Dust(f, 1.5, 16), color='r', label='B')
axis.loglog(f, Dust(f, 1.5, 21), color='b', label='C')
axis.loglog(f, Dust(f, 1.5, 26), color='y', label='D')
axis.loglog(f, Dust(f, 2.0, 21), color='k', label='E')
axis.set_title("Thermal Dust")
axis.set_xlabel("Frequency (GHz)")
axis.set_ylabel(r"Brightness temperature ($\mu$K)")
axis.legend()
axis.plot()
```
